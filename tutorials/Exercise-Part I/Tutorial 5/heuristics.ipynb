```python



# Function to calculate makespan
def calculate_makespan(sequence, jobs):
    n_jobs = len(sequence)
    n_machines = len(jobs[sequence[0]])
    completion_times = np.zeros((n_jobs, n_machines))

    for i, job_id in enumerate(sequence):
        for m in range(n_machines):
            if i == 0 and m == 0:
                completion_times[i, m] = jobs[job_id][m]
            elif i == 0:
                completion_times[i, m] = completion_times[i, m-1] + jobs[job_id][m]
            elif m == 0:
                completion_times[i, m] = completion_times[i-1, m] + jobs[job_id][m]
            else:
                completion_times[i, m] = max(completion_times[i-1, m], completion_times[i, m-1]) + jobs[job_id][m]
    return completion_times[-1, -1]

# SPT: sort jobs by total processing time (ascending)
spt_sequence = sorted(jobs_original, key=lambda j: sum(jobs_original[j]))
makespan_spt = calculate_makespan(spt_sequence, jobs_original)

print("SPT Job Sequence (Original):", spt_sequence)
print("Makespan (Original):", makespan_spt)
```

    Job 4: [13, 22, 14, 13]

    Job 3: [23, 42, 27, 6]

We build a completion matrix: completion[i][m]
	M1	M2	M3	M4
Job 4

🔹 Job 4 (i = 0)

Since it's the first job, it runs continuously:

completion[0, 0] = 13                   # M1


completion[0, 1] = 13 + 22 = 35         # M2

completion[0, 2] = 35 + 14 = 49         # M3

completion[0, 3] = 49 + 13 = 62         # M4

🔹 Job 3 (i = 1)

This job must wait:

    For M1 to be free (after Job 4 finishes on M1),

    And for itself to finish on the previous machine.

completion[1, 0] = 13 + 23 = 36                   # M1

completion[1, 1] = max(35, 36) + 42 = 78          # M2

completion[1, 2] = max(49, 78) + 27 = 105         # M3

completion[1, 3] = max(62, 105) + 6 = 111         # M4

 Final Makespan = 111 (after 2 jobs)

In full execution, the matrix is filled row-by-row using:

completion[i, m] = max(completion[i-1, m], completion[i, m-1]) + proc_time

The final makespan is:

completion[-1, -1] = value in bottom-right corner = finish time of last job on last machine

```python
from gurobipy import Model, GRB
import itertools

# --- Job instance ---
# jobs = {
#     1: [31, 41, 25, 30],
#     2: [19, 55, 3, 34],
#     3: [23, 42, 27, 6],
#     4: [13, 22, 14, 13],
#     5: [33, 5, 57, 19]
# }
# To test on extended instance, uncomment below
jobs = jobs_extended

# --- Problem parameters ---
n = len(jobs)
m = len(next(iter(jobs.values())))
job_ids = list(jobs.keys())

# --- Compute Big M as: n * max(p_ik) ---
max_p = max(max(times) for times in jobs.values())
M = n * max_p
print(f"Computed Big M = {M}")

# --- Gurobi model ---
model = Model("Manne_PFSP")

# Variables
C = model.addVars(job_ids, range(m), name="C", vtype=GRB.CONTINUOUS)
z = model.addVars(itertools.combinations(job_ids, 2), vtype=GRB.BINARY, name="z")
Cmax = model.addVar(name="Cmax", vtype=GRB.CONTINUOUS)

# --- Constraints ---

# 1. Job flow across machines
for i in job_ids:
    for k in range(1, m):
        model.addConstr(C[i, k] >= C[i, k-1] + jobs[i][k], name=f"flow_{i}_{k}")

# 2. Precedence constraints
for i, j in itertools.combinations(job_ids, 2):
    for k in range(m):
        model.addConstr(C[j, k] >= C[i, k] + jobs[i][k] - M * (1 - z[i, j]), name=f"prec_{i}_{j}_{k}")
        model.addConstr(C[i, k] >= C[j, k] + jobs[j][k] - M * z[i, j], name=f"prec_{j}_{i}_{k}")

# 3. Makespan definition
for i in job_ids:
    model.addConstr(Cmax >= C[i, m-1], name=f"Cmax_{i}")

# Objective: minimize makespan
model.setObjective(Cmax, GRB.MINIMIZE)

# Solve
model.optimize()

# --- Output ---
if model.status == GRB.OPTIMAL:
    print(f"\n Optimal Makespan (Cmax): {Cmax.X:.1f}")
    for i in job_ids:
        for k in range(m):
            print(f"Job {i} on Machine {k+1}: Completion = {C[i, k].X:.1f}")
else:
    print("No optimal solution found.")
```
