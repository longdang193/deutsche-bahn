# Example 3 - Set representation
## Task 
Solve the following mathematical optimization problem:

$$
\begin{equation*}
	\textrm{max } G =  \sum_{j=1}^{J}g_jx_j
\end{equation*}
$$

$$
\begin{align*}
	&& \textrm{subject to constraints: }  \sum_{j=1}^{J}a_{ij}x_j &\leq T_i && \textrm{for } i = 1,2,\dots,I \\
	&& x_j &\geq 0 &&  \textrm{for } j = 1,2,\dots,J
\end{align*}
$$

with following data:

$$
\begin{equation*}
    g = 
    \begin{pmatrix}
    10 \\
    20 \\
    5
    \end{pmatrix}, 
        a = 
    \begin{pmatrix}
    6 & 2 & 4 \\
    10 & 10 & 8 \\
	1 & 4 & 6
    \end{pmatrix},
    T = 
    \begin{pmatrix}
    480 \\
    1000 \\
    280
    \end{pmatrix}
\end{equation*}
$$

```python
import gurobipy as gp
from gurobipy import GRB
```

## Definition of sets

```python
I_max = 3
J_max = 3

I = range(I_max)
J = range(J_max)
```

## Entering the parameters
$$
\begin{equation*}
    g = 
    \begin{pmatrix}
    10 \\
    20 \\
    5
    \end{pmatrix}
\end{equation*}
$$

```python
g = [10, 20, 5]
```

$$
\begin{equation*}
    a = 
    \begin{pmatrix}
    6 & 2 & 4\\
    10 & 10 & 8\\
	1 & 4 & 6
    \end{pmatrix}
\end{equation*}
$$

```python
a = [[6, 2, 4],
    [10, 10, 8],
    [1, 4, 6]]
```

$$
\begin{equation*}
    T = 
    \begin{pmatrix}
    480 \\
    1000 \\
    280
    \end{pmatrix}
\end{equation*}
$$

```python
T = [480, 1000, 280]
```

## Initialization of the model

```python
m = gp.Model()
```

## Initialization of the variables
$$ x_j \geq 0 \textrm{ for } j = 1,2\dots,J $$
Detailed version:

```python
x={}
for j in J:
    x[j] = m.addVar(obj=g[j], name="x"+str(j))
```

## Defining the objective function

$$ \textrm{max } G =  \sum_{j=1}^{J}g_jx_j $$

This is already defined here via the objective function coefficients 'obj' of the variables. However, it can also be defined explicitly.


```python
m.modelSense = GRB.MAXIMIZE
```

## Adding the constraints

$$ \sum_{j=1}^{J}a_{ij}x_j \leq T_i  \textrm{ for } i = 1,2,\dots,I $$

Detailed version:

```python
for i in I:
    m.addConstr(gp.quicksum(a[i][j] * x[j] for j in J) <= T[i], name="c"+str(i))
```

## Optimization of the model

```python
m.optimize()
```

## Result output

```python
m.printAttr(GRB.Attr.X)
```

```python
m.printAttr(GRB.Attr.ObjVal)
```

 
## Sensitivity analysis
Determine the reduced costs of all decision variables.


```python
for var in m.getVars():
    print(var.getAttr(GRB.Attr.VarName), var.getAttr(GRB.Attr.RC))
```

Output the slack of the first constraint.

```python
c0 = m.getConstrByName("c0")
c0.getAttr(GRB.Attr.Slack)
```

 
Enter the opportunity costs of the second and third constraint.

```python
c1 = m.getConstrByName("c1")
c2 = m.getConstrByName("c2")

for i in [c1,c2]:
    print(i.getAttr(GRB.Attr.ConstrName), i.getAttr(GRB.Attr.Pi))
```
