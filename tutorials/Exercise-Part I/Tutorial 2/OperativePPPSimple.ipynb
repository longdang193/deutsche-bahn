# Example 3

## Task

Solve the following mathematical optimization problem:
$$
	\begin{align*}
		\textrm{max } 10x_1 + 20x_2 &= G \\
		& \\
		\textrm{in the following: } 6x_1 + 2x_2 &\leq 480 \\
		10x_1 + 10x_2 &\leq 1000 \\
		1x_1 + 4x_2 &\leq 280 \\
		x_1, x_2 &\leq 0	
	\end{align*}
$$

## Initialization of the Gurobi model

Loading the modules:

```python
import gurobipy as gp
from gurobipy import GRB
```

Creating the model:

```python
m = gp.Model()
```

## Initialization of the variables

$$ x_1, x_2 \geq 0 $$

```python
x1 = m.addVar(vtype=GRB.CONTINUOUS, obj=10, lb=0, name="x_1")
x2 = m.addVar(vtype=GRB.CONTINUOUS, obj=20, lb=0, name="x_2")
```

## Definition of the objective function

$$ \textrm{max } 10x_1 + 20x_2 = G $$

```python
m.modelSense = GRB.MAXIMIZE
# m.setObjective = (10x1 + 20x2, GRB.MAXIMIZE)
```

## Add the constraints
$$
\begin{align*}
	6x_1 + 2x_2 &\leq 480 \\
	10x_1 + 10x_2 &\leq 1000 \\
	1x_1 + 4x_2 &\leq 280
\end{align*}
$$

```python
m.addConstr(6*x1 + 2*x2 <= 480,"nb_I")
m.addConstr(10*x1 + 10*x2 <= 1000, "nb_II")
m.addConstr(x1 + 4*x2 <= 280, "nb_III")
```

## Optimization of the model

```python
m.optimize()
```

## Result output

```python
m.printAttr(GRB.Attr.X)
```

```python
m.printAttr(GRB.Attr.ObjVal)
```
