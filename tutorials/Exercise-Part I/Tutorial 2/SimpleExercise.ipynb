# Example 2 - Leisure activities

## Task

Solve the following mathematical optimization problem:
$$
	\begin{align*}
		\textrm{max } x + y +2z& \\
		& \\
		\textrm{in the following: } 0.5x + y + 2 z &\leq 3 \\
		x + y &= 1 \\[1mm]
		x,y,z &\in \{0,1\}		
	\end{align*}
$$

## Initialization of the Gurobi model

Before the components and functions of Gurobi can be used, the module must be loaded. 

```python
import gurobipy as gp
from gurobipy import GRB
```

The first step is to create an empty model with the name m.

```python
m = gp.Model()
```

## Initialization of the decision variables

The decision variables x, y and z are added to the model m step by step.
\begin{equation}
		x,y,z \in \{ 0,1 \}	
\end{equation}

```python
x = m.addVar(vtype=GRB.BINARY, name="x")
y = m.addVar(vtype=GRB.BINARY, name="y")
z = m.addVar(vtype=GRB.BINARY, name="z")
```

## Definition of the target function

The previously defined variables can be used to define the target function. The target function consists of a mathematical term and the optimization direction (min / max).
$$
\begin{equation*}
    \textrm{max } x + y +2z \\
\end{equation*}
$$

```python
m.setObjective(x + y + 2*z, GRB.MAXIMIZE)
```

## Adding the constraints

Similarly, the constraints can be added.
$$
	\begin{align*}
		0.5x + y + 2 z &\leq 3 \\
		x + y &= 1 \\
		
	\end{align*}
$$

```python
m.addConstr(0.5*x + y + 2*z <= 3)
m.addConstr(x + y == 1)
```

## Optimization of the model

Solve the model:

```python
m.optimize()
```

## Result output

With the command `m.printAttr('X')` the solution values $(>0)$ of all variables are displayed:

```python
m.printAttr('X')
```

The command `m.printAttr('ObjVal')` displays the target function value:

```python
m.printAttr('ObjVal')
```
