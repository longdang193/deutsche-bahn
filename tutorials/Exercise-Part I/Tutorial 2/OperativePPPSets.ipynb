# Example 3 - Set representation

## Task 

Solve the following mathematical optimization problem:
$$
\begin{equation*}
	\textrm{max } G = \sum_{j=1}^{J}g_jx_j
\end{equation*}
$$

$$
\begin{align*}
	&& \textrm{in the following: }  \sum_{j=1}^{J}a_{ij}x_j &\leq T_i && \textrm{for } i = 1,2,\dots,I \\
	&& x_j &\geq 0 && \textrm{for } j = 1,2,\dots,J
\end{align*}
$$

with the following data:
$$
\begin{equation*}
    g = 
    \begin{pmatrix}
    10 \\
    20
    \end{pmatrix}, 
        a = 
    \begin{pmatrix}
    6 & 2 \\
    10 & 10 \\
	1 & 4
    \end{pmatrix},
    T = 
    \begin{pmatrix}
    480 \\
    1000 \\
    280
    \end{pmatrix}
\end{equation*}
$$

```python
import gurobipy as gp
from gurobipy import GRB
```

## Definition of the sets 

```python
I_max = 3
J_max = 2

I = range(I_max)
J = range(J_max)
```

## Entering the Parameter
$$
\begin{equation*}
    g = 
    \begin{pmatrix}
    10 \\
    20
    \end{pmatrix}
\end{equation*}
$$

```python
g = [10, 20]
```

$$
\begin{equation*}
    a = 
    \begin{pmatrix}
    6 & 2 \\
    10 & 10 \\
	1 & 4
    \end{pmatrix}
\end{equation*}
$$

```python
a = [[6, 2],
    [10, 10],
    [1, 4]]
```

$$
\begin{equation*}
    T = 
    \begin{pmatrix}
    480 \\
    1000 \\
    280
    \end{pmatrix}
\end{equation*}
$$

```python
T = [480, 1000, 280]
```

## Initialising the model

```python
m = gp.Model()
```

## Initialization of the variables
$$ x_j \geq 0 \textrm{ for } j = 1,2\dots,J $$
Detailed version:

```python
x={}
for j in J:
    x[j] = m.addVar(obj=g[j], name="x"+str(j))
```

Summary:

```python
#x = m.addVars(J_max, obj=g, name="x")
```

## Definition of the objective function

$$ \textrm{max } G = \sum_{j=1}^{J}g_jx_j $$

This is already defined here via the objective function coefficients `obj` of the variables. However, it can also be defined explicitly.

```python
m.modelSense = GRB.MAXIMIZE
#m.setObjective(gp.quicksum(g[j] * x[j] for j in J),GRB.MAXIMIZE)
```

## Adding the constraints

$$ \sum_{j=1}^{J}a_{ij}x_j \leq T_i \textrm{ for } i = 1,2,\dots,I $$

Detailed version:

```python
for i in I:
    m.addConstr(gp.quicksum(a[i][j] * x[j] for j in J) <= T[i], name="c"+str(i))
```

Short version: 

```python
#m.addConstrs((gp.quicksum(a[i][j] * x[j] for j in J) <= T[i] for i in I), name="c")
```

## Optimizing the model

```python
m.optimize()
```

## Results output

```python
m.printAttr(GRB.Attr.X)
```

```python
m.printAttr(GRB.Attr.ObjVal)
```
