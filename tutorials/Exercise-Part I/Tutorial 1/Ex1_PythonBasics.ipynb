## Lists
---
A list is initialized by `[]` or `list()`.

```python
myList = [1, 20, 3]
```

Lists can be sorted in ascending order with `list.sort()`. If you want to sort in descending order, this can be defined using the parameter `reverse=True`.

```python
myList.sort()   # myList.sort(reverse=True)
myList
```

An element `a` can be added with `list.append(a)`.

```python
myList.append("text")
myList
```

Nested lists can also be created. The following matrix is to be generated as a nested list:
$$\begin{equation}
    B = 
    \begin{pmatrix}
    1 & 0 \\
    3 & 5
    \end{pmatrix}
\end{equation}$$

```python
mySecondList = [[1, 0], [3, 5]]
mySecondList
```

Lists can be added to existing lists.

```python
myList.extend(mySecondList)
myList
```

The length of a list can be output.

```python
len(myList)
```

Elements can be queried individually or as a range of several elements via indices.

```python
myList[0] # 1. Element
```

```python
myList[2:5] # 3. until 5. Element
```

Elements can also be deleted from a list.

```python
del myList[0]
myList
```

## Conditional statements
---
Conditional statements consist of a query and an If block that is executed if the statement is true.

```python
value = 1
```

```python
if type(value) == int or type(value) == float:
    print("It's a number!")
elif type(value) == str:
    print("It's a string!")
else:
    print("It is neither a number nor a string!")
```

### Example:
Test whether a number is positive, negative or zero.

```python
num = 1
```

nested if statement:

```python
if num >= 0:
    if num == 0:
        print("Null")
    else:
        print("Positive number")
else:
    print("Negative number")
```

## Loops
---
Loops consist of an iterator, also known as a run variable, and an enumeration that is iterated over.

```python
for item in myList:
    print(item)
```

```python
for (index, item) in enumerate(myList):
    print(index, ":", item)
```

A list does not necessarily have to be generated beforehand. You can also work with auxiliary constructs such as `range`.

### Example:
Multiplication table (from 1 to 10) in Python with a given base

```python
basis = 12
```

Iterate 10 times from i = 1 to 10.

```python
for i in range(1, 11):
   print(f"{basis} x {i} = {basis*i}")
```

Loops can also be used to create enumerations. Each letter of the following word is to be added to a list.

```python
word = "Human"
```

Long Form:

```python
h_letters = []

for letter in word:
    h_letters.append(letter)

print(h_letters)
```

Short Form:

```python
word = "Human"
h_letters = [ letter for letter in word ]
print(h_letters)
```

## Writing files
---
Overwriting write permissions can be assigned:

```python
with open("myFirstTextFile.txt", "w", encoding="utf-8") as f:
    f.write("Hello world! \n")
    f.write("\tIndentations are possible.\n")
```

Additional write permissions can be assigned:

```python
with open("myFirstTextFile.txt", "a", encoding="utf-8") as f:
    f.write("1+1=2\n")
```

The default authorization for `open()` is read authorization only:

```python
with open("myFirstTextFile.txt", encoding="utf-8") as f:
    #print(f.read(4))
    #print(f.readline())
    print(f.read())
```

## Importing csv files
---
To be able to read csv files, the `csv` module must be loaded.

```python
import csv
```

Importing a csv file works in the same way as importing a txt file.

```python
with open("MitarbeiterDaten.csv", encoding="utf-8") as csv_file:
    csv_reader = csv.DictReader(csv_file)    
    for index,row in enumerate(csv_reader):
        if index == 0:
            print(f"The column names are: {list(row.keys())}")            
        print(f"\t {row['Name']} works in the {row['Abteilung']} department and was born in {row['Geburtstagsmonat']}.")    
```

**Problem**: With large cost matrices, it is inconvenient to query each column individually.

If the keys of a dictionary are not used, it does not have to be saved as a dictionary. The more general form for reading is the `csv.reader()`. Since the first line with column names is not necessary, it can be skipped. To do this, `next()` is used and the first line is output for checking. The row indices as the first element in each row can be skipped directly using indexing.

```python
costs=[]
with open("KostenMatrix.csv", encoding="utf-8") as csv_file:
    csv_reader = csv.reader(csv_file)
    header = next(csv_reader)
    print(header)
    for row in csv_reader:
        rowAsInt = [int(item) for item in row[1:]]
        costs.append(rowAsInt)
costs
```
